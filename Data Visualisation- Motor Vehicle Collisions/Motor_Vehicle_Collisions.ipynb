## Filter The Data to 2018 to 2021

```python
# Convert 'CRASH DATE' to datetime format
collision_data['CRASH DATE'] = pd.to_datetime(collision_data['CRASH DATE'], errors='coerce')

# Filter data for the years 2018 to 2021
data_filtered = collision_data[(collision_data['CRASH DATE'] >= '2018-01-01') & (collision_data['CRASH DATE'] <= '2021-12-31')]

# Save the filtered data to a new CSV file
data_filtered.to_csv('motor_vehicle_collisions_2018_2021.csv', index=False)
```

Handling Missing values

```python
# Drop rows with missing  'BOROUGH', 'ZIP CODE', 'LATITUDE', 'LONGITUDE' and  'LOCATION'
data_cleaned1 = data_filtered.dropna(subset=['BOROUGH', 'ZIP CODE','LATITUDE', 'LONGITUDE', 'LOCATION'])

# Save the cleaned data to a new CSV file
data_cleaned1.to_csv('motor_vehicle_collisions_2018_2021_cleaned1.csv', index=False)
```

```python
# Check for remaining missing values
data_cleaned1.isnull().sum()
```

```python
data_cleaned1
```

```python
#Handle 0 values in the lattitude and longitude columns
# Filter rows where latitude or longitude equals 0

data_cleaned1[(data_cleaned1['LATITUDE'] == 0) | (data_cleaned1['LONGITUDE'] == 0)]
```

**Handling the values of 0 for the column Latitude and Longitude**

If the 0 values in the column latitude and longitude is representing the zero coordinates in the map, then the other column values  like borough, etc should be the same. But, here in this case it is not.

From that we can get to know that here these 0 values shows the missing values of those columns

```python
# Filter rows where latitude or longitude equals 0
zero_records_indices = data_cleaned1[(data_cleaned1['LATITUDE'] == 0) | (data_cleaned1['LONGITUDE'] == 0)].index

# Drop rows with latitude or longitude as 0
data_cleaned2 = data_cleaned1.drop(zero_records_indices)

# Optional: Reset index if needed
data_cleaned2.reset_index(drop=True, inplace=True)

# Print summary
print(f"Dropped {len(zero_records_indices)} rows where latitude or longitude equals 0.")
print(f"Dataset size after removing rows with latitude or longitude as 0: {len(data_cleaned2)} rows")

# Save cleaned dataset if needed
data_cleaned2.to_csv('motor_vehicle_collisions_2018_2021_cleaned2.csv', index=False)
```

```python
data_cleaned2
```

```python
data_cleaned2.isnull().sum()
```

```python
# Handle missing values in other columns
data_cleaned2['NUMBER OF PERSONS INJURED'].fillna(0, inplace=True)
data_cleaned2['NUMBER OF PERSONS KILLED'].fillna(0, inplace=True)
data_cleaned2['CONTRIBUTING FACTOR VEHICLE 1'].fillna('Unspecified', inplace=True)
data_cleaned2['CONTRIBUTING FACTOR VEHICLE 2'].fillna('Unspecified', inplace=True)
data_cleaned2['CONTRIBUTING FACTOR VEHICLE 3'].fillna('Unspecified', inplace=True)
data_cleaned2['CONTRIBUTING FACTOR VEHICLE 4'].fillna('Unspecified', inplace=True)
data_cleaned2['CONTRIBUTING FACTOR VEHICLE 5'].fillna('Unspecified', inplace=True)

data_cleaned2['VEHICLE TYPE CODE 1'].fillna('Unspecified', inplace=True)
data_cleaned2['VEHICLE TYPE CODE 2'].fillna('Unspecified', inplace=True)
data_cleaned2['VEHICLE TYPE CODE 3'].fillna('Unspecified', inplace=True)
data_cleaned2['VEHICLE TYPE CODE 4'].fillna('Unspecified', inplace=True)
data_cleaned2['VEHICLE TYPE CODE 5'].fillna('Unspecified', inplace=True)

data_cleaned2['ON STREET NAME'].fillna('Unspecified', inplace=True)
data_cleaned2['CROSS STREET NAME'].fillna('Unspecified', inplace=True)
data_cleaned2['OFF STREET NAME'].fillna('Unspecified', inplace=True)
```

## Data Visualisation

#### Number of Motor Vehicle Collisions per year - 2018 to 2021

```python
# Add a YEAR column
data_cleaned2['YEAR'] = data_cleaned2['CRASH DATE'].dt.year

# Aggregate data by year
collisions_per_year = data_cleaned2.groupby('YEAR').size().reset_index(name='Collisions')

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='YEAR', y='Collisions', data=collisions_per_year, palette='viridis')
plt.title('Motor Vehicle Collisions by Year (2018-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Collisions')
plt.tight_layout()
plt.show()
```

```python
# Correcting the Future warning
# Add a YEAR column
# data_cleaned2['YEAR'] = data_cleaned2['CRASH DATE'].dt.year

# Aggregate data by year
collisions_per_year = data_cleaned2.groupby('YEAR').size().reset_index(name='Collisions')

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='YEAR', y='Collisions', data=collisions_per_year, palette='viridis', hue='YEAR', dodge=False, legend=False)
plt.title('Motor Vehicle Collisions by Year (2018-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Collisions')
plt.tight_layout()
plt.show()
```

```python
#Showing the collision values per year in the bar chart
# Add a YEAR column
# data_cleaned2['YEAR'] = data_cleaned2['CRASH DATE'].dt.year

# Aggregate data by year
collisions_per_year = data_cleaned2.groupby('YEAR').size().reset_index(name='Collisions')

# Create the bar chart
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='YEAR', y='Collisions', data=collisions_per_year, palette='viridis')

# Annotate the bars with the collision counts
for index, row in collisions_per_year.iterrows():
    barplot.text(row['YEAR'] - collisions_per_year['YEAR'].min(), row['Collisions'], round(row['Collisions'], 2), color='black', ha="center")

plt.title('Motor Vehicle Collisions by Year (2018-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Collisions')
plt.tight_layout()
plt.show()
```

#### Number of Motor Vehicle Collisions per Year across different Boroughs

```python
# Aggregate data by year and borough
collisions_per_year_borough = data_cleaned2.groupby(['YEAR', 'BOROUGH']).size().reset_index(name='Collisions')

# Create the grouped bar chart
plt.figure(figsize=(14, 8))
barplot = sns.barplot(x='YEAR', y='Collisions', hue='BOROUGH', data=collisions_per_year_borough, palette='viridis')

# Annotate the bars with the collision counts
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9),
                     textcoords = 'offset points')

plt.title('Number of Motor Vehicle Collisions per Year across Different Boroughs (2018-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Collisions')
plt.legend(title='Borough')
plt.tight_layout()
plt.show()

```

#### Number of Injuries and Fatalities per year

```python
# Aggregate data by year and sum the injuries and fatalities
collisions_per_year = data_cleaned2.groupby('YEAR').agg({
    'NUMBER OF PERSONS INJURED': 'sum',
    'NUMBER OF PERSONS KILLED': 'sum',
    'NUMBER OF PEDESTRIANS INJURED': 'sum',
    'NUMBER OF PEDESTRIANS KILLED': 'sum',
    'NUMBER OF CYCLIST INJURED': 'sum',
    'NUMBER OF CYCLIST KILLED': 'sum',
    'NUMBER OF MOTORIST INJURED': 'sum',
    'NUMBER OF MOTORIST KILLED': 'sum'
}).reset_index()

# Melt the dataframe to long format for easier plotting
collisions_per_year_melted = pd.melt(collisions_per_year, id_vars=['YEAR'], var_name='Category', value_name='Count')

# Create the grouped bar chart
plt.figure(figsize=(14, 8))
barplot = sns.barplot(x='YEAR', y='Count', hue='Category', data=collisions_per_year_melted, palette='viridis')

# Annotate the bars with the counts
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9),
                     textcoords = 'offset points')

plt.title('Number of Injuries and Fatalities per Year (2018-2021)')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Category', loc='upper center')
plt.tight_layout()
plt.show()
```

#### Number of Injuries and Fatalities across different Boroughs

```python
# Aggregate data by borough and sum injuries and fatalities
collisions_per_borough = data_cleaned2.groupby('BOROUGH').agg({
    'NUMBER OF PERSONS INJURED': 'sum',
    'NUMBER OF PERSONS KILLED': 'sum',
    'NUMBER OF PEDESTRIANS INJURED': 'sum',
    'NUMBER OF PEDESTRIANS KILLED': 'sum',
    'NUMBER OF CYCLIST INJURED': 'sum',
    'NUMBER OF CYCLIST KILLED': 'sum',
    'NUMBER OF MOTORIST INJURED': 'sum',
    'NUMBER OF MOTORIST KILLED': 'sum'
}).reset_index()

# Melt the dataframe to long format for easier plotting
collisions_per_borough_melted = pd.melt(collisions_per_borough, id_vars='BOROUGH',
                                        var_name='Category', value_name='Count')

# Create the grouped bar chart
plt.figure(figsize=(14, 8))
sns.barplot(x='BOROUGH', y='Count', hue='Category', data=collisions_per_borough_melted, palette='viridis')

# Annotate the bars with the counts
for p in plt.gca().patches:
    plt.gca().annotate(format(p.get_height(), '.0f'),
                       (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center',
                       xytext=(0, 9),
                       textcoords='offset points')

plt.title('Number of Injuries and Fatalities across Different Boroughs')
plt.xlabel('Borough')
plt.ylabel('Count')
plt.legend(title='Category', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Number of Injuries and fatalities across ZipCode

```python
# Aggregate data by zip code and sum injuries and fatalities
collisions_per_zipcode = data_cleaned2.groupby('ZIP CODE').agg({
    'NUMBER OF PERSONS INJURED': 'sum',
    'NUMBER OF PERSONS KILLED': 'sum',
    'NUMBER OF PEDESTRIANS INJURED': 'sum',
    'NUMBER OF PEDESTRIANS KILLED': 'sum',
    'NUMBER OF CYCLIST INJURED': 'sum',
    'NUMBER OF CYCLIST KILLED': 'sum',
    'NUMBER OF MOTORIST INJURED': 'sum',
    'NUMBER OF MOTORIST KILLED': 'sum'
}).reset_index()

# Melt the dataframe to long format for easier plotting
collisions_per_zipcode_melted = pd.melt(collisions_per_zipcode, id_vars='ZIP CODE',
                                        var_name='Category', value_name='Count')

# Create the grouped bar chart
plt.figure(figsize=(16, 8))
plt.bar(collisions_per_zipcode_melted['ZIP CODE'], collisions_per_zipcode_melted['Count'],
        color='skyblue')

plt.xlabel('ZIP Code')
plt.ylabel('Count')
plt.title('Number of Injuries and Fatalities across ZIP Codes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Aggregate data by ZIP code and sum injuries and fatalities
collisions_per_zipcode = data_cleaned2.groupby('ZIP CODE').agg({
    'NUMBER OF PERSONS INJURED': 'sum',
    'NUMBER OF PERSONS KILLED': 'sum',
    'NUMBER OF PEDESTRIANS INJURED': 'sum',
    'NUMBER OF PEDESTRIANS KILLED': 'sum',
    'NUMBER OF CYCLIST INJURED': 'sum',
    'NUMBER OF CYCLIST KILLED': 'sum',
    'NUMBER OF MOTORIST INJURED': 'sum',
    'NUMBER OF MOTORIST KILLED': 'sum'
}).reset_index()

# Calculate total injuries and fatalities
collisions_per_zipcode['Total Injuries and Fatalities'] = (
    collisions_per_zipcode['NUMBER OF PERSONS INJURED'] +
    collisions_per_zipcode['NUMBER OF PEDESTRIANS INJURED'] +
    collisions_per_zipcode['NUMBER OF CYCLIST INJURED'] +
    collisions_per_zipcode['NUMBER OF MOTORIST INJURED'] +
    collisions_per_zipcode['NUMBER OF PERSONS KILLED'] +
    collisions_per_zipcode['NUMBER OF PEDESTRIANS KILLED'] +
    collisions_per_zipcode['NUMBER OF CYCLIST KILLED'] +
    collisions_per_zipcode['NUMBER OF MOTORIST KILLED']
)

# Find the ZIP code with the highest number of injuries and fatalities
max_injuries_fatalities_zip = collisions_per_zipcode.loc[
    collisions_per_zipcode['Total Injuries and Fatalities'].idxmax()]

print("ZIP code with the highest number of injuries and fatalities:")
print(max_injuries_fatalities_zip['ZIP CODE'])
```

Number of Collisions Per ZipCode

```python
# Count collisions by ZIP code
collisions_per_zipcode = data_cleaned2['ZIP CODE'].value_counts().reset_index()
collisions_per_zipcode.columns = ['ZIP CODE', 'Collisions']

# Sort by ZIP code for better visualization (optional)
collisions_per_zipcode = collisions_per_zipcode.sort_values(by='ZIP CODE')

# Create the bar plot
plt.figure(figsize=(16, 8))
plt.bar(collisions_per_zipcode['ZIP CODE'], collisions_per_zipcode['Collisions'], color='skyblue')

plt.xlabel('ZIP Code')
plt.ylabel('Number of Collisions')
plt.title('Number of Motor Vehicle Collisions across ZIP Codes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Count collisions by ZIP code
collisions_per_zipcode = data_cleaned2['ZIP CODE'].value_counts().reset_index()
collisions_per_zipcode.columns = ['ZIP CODE', 'Collisions']

# Find the ZIP code with the highest number of collisions
max_collisions_zip = collisions_per_zipcode.loc[collisions_per_zipcode['Collisions'].idxmax()]

# Find the ZIP code with the lowest number of collisions
min_collisions_zip = collisions_per_zipcode.loc[collisions_per_zipcode['Collisions'].idxmin()]

print("ZIP code with the highest number of collisions:")
print(max_collisions_zip)

print("ZIP code with the lowest number of collisions:")
print(min_collisions_zip)
```
